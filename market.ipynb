```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing data 
data=pd.read_csv("Market2.csv")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
# selecting NYA rows 
df = df.loc[df['Index'] == 'NYA']
df
```

```python
# converting format of date to pandas format
df['Date']=pd.to_datetime(df.Date)
df
```

```python
############ step1 : finding noise with scatter plot between columns ###########
```

```python
#plot Date & Open price  before dropping missing values to find the location of noise 
fig = plt.figure(figsize=(15,8),dpi=80)
fig.suptitle("date & Adj Close", fontsize=20)
plt.scatter(df['Date'],df['Adj Close'])
plt.xlabel("Date",fontsize=20)
plt.ylabel("Adj Close",fontsize=20)
plt.grid()
plt.show()
```

```python
#finding location of noise
mask = (df['Adj Close']>5996)&(df['Adj Close']<6500)
df100 = df.loc[mask]
mask2 =(df100['Date']<'1970-01-01')
print(df100.loc[mask2])
```

```python
#plot Date & Open price  before dropping missing values to find the location of noise 
fig = plt.figure(figsize=(15,8),dpi=80)
fig.suptitle('date & open price', fontsize=20)
plt.scatter(df['Date'],df['Open'])
plt.xlabel("Date",fontsize=20)
plt.ylabel("Open",fontsize=20)
plt.grid()
plt.show()
```

```python
#finding location of noise
mask = (df['Open']>5500)&(df['Open']<6000)
df100 = df.loc[mask]
mask2 =(df100['Date']<'1970-01-01')
print(df100.loc[mask2])
```

```python
#plot Date & High price  before dropping missing values to find the location of noise 
fig = plt.figure(figsize=(15,8),dpi=80)
fig.suptitle('date & High price', fontsize=20)
plt.scatter(df['Date'],df['High'])
plt.xlabel("Date",fontsize=20)
plt.ylabel("High",fontsize=20)
plt.grid()
plt.show()
```

```python
#finding location of noise
mask = (df['High']>5900)&(df['High']<6200)
df100 = df.loc[mask]
mask2 =(df100['Date']<'1970-01-01')
print(df100.loc[mask2])
```

```python
#plot Date & Low price  before dropping missing values to find the location of noise 
fig = plt.figure(figsize=(15,8),dpi=80)
fig.suptitle('date & Low price', fontsize=20)
plt.scatter(df['Date'],df['Low'])
plt.xlabel("Date",fontsize=20)
plt.ylabel("Low",fontsize=20)
plt.grid()
plt.show()
```

```python
# there is not any noise in the above plot (Date & Low price)
```

```python
#plot Date & Close price  before dropping missing values to find the location of noise 
fig =plt.figure(figsize=(15,8),dpi=80)
fig.suptitle('date & Close price', fontsize=20)
plt.scatter(df['Date'],df['Close'])
plt.xlabel("Date",fontsize=20)
plt.ylabel("Close",fontsize=20)
plt.grid()
plt.show()
```

```python
# there is not any noise in the above plot (Date & Close price)
```

```python
############ step2 : making a new dataframe without noise ##########
```

```python
# we found location of noise in below indexes of first dataframe so we clean them before dropping missing values:
#row 829
#row 831
#row 833
#row 852
```

```python
#deleting specified rows at index values : 829 , 831 , 833 , 852
df_withoutnoise = df.drop(labels = [829,831,833,852], axis =0)
df_withoutnoise
```

```python
################# testing dataframe without noise ################
```

```python
#plot Date & Adj Close price  before dropping missing values and after dropping noise 
#now you can see there is not any noise in this plot
fig = plt.figure(figsize=(15,8),dpi=80)
plt.scatter(df_withoutnoise['Date'],df_withoutnoise['Adj Close'])
fig.suptitle("date & Adj Close", fontsize=20)
plt.xlabel("Date",fontsize=20)
plt.ylabel("Adj Close",fontsize=20)
plt.grid()
plt.show()
```

```python
#plot Date & Open price  before dropping missing values and after dropping noise 
#now you can see there is not any noise in this plot
fig = plt.figure(figsize=(15,8),dpi=80)
plt.scatter(df_withoutnoise['Date'],df_withoutnoise['Open'])
fig.suptitle("date & Open", fontsize=20)
plt.xlabel("Date",fontsize=20)
plt.ylabel("Open",fontsize=20)
plt.grid()
plt.show()
```

```python
#plot Date & High price  before dropping missing values and after dropping noise 
#now you can see there is not any noise in this plot

fig = plt.figure(figsize=(15,8),dpi=80)
plt.scatter(df_withoutnoise['Date'],df_withoutnoise['High'])
fig.suptitle("date & High", fontsize=20)
plt.xlabel("Date",fontsize=20)
plt.ylabel("High",fontsize=20)
plt.grid()
plt.show()
```

```python
#describing dataframe without noise
#now you can see 4 rows are dropped from each column
df_withoutnoise.describe()
```

```python
################## step3 : dropping missing values #####################
```

```python
#dropping missing values
#making final datafrme without noise and missing values
df_final = df_withoutnoise.dropna()
df_final
```

```python
########################## plot final dataframe #########################
```

```python
#Date & Adj Close
fig = plt.figure(figsize=(15,8),dpi=80)
plt.plot(df_final['Date'],df_final['Adj Close'])
fig.suptitle("date & Adj Close", fontsize=20)
plt.xlabel("Date",fontsize=20)
plt.ylabel("Adj Close",fontsize=20)
plt.grid()
plt.show()
```

```python
#Date & Open
fig = plt.figure(figsize=(15,8),dpi=80)
plt.plot(df_final['Date'],df_final['Open'])
fig.suptitle("date & Open", fontsize=20)
plt.xlabel("Date",fontsize=20)
plt.ylabel("Open",fontsize=20)
plt.grid()
plt.show()
```

```python
#Date & High
fig = plt.figure(figsize=(15,8),dpi=80)
plt.plot(df_final['Date'],df_final['High'])
fig.suptitle("date & High", fontsize=20)
plt.xlabel("Date",fontsize=20)
plt.ylabel("High",fontsize=20)
plt.grid()
plt.show()
```
